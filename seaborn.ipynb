## Seaborn

Seaborn is a visualization library built on top of Matplotlib, with easier one-liners

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configurations
%matplotlib inline
%reload_ext autoreload
%autoreload 2
```

```python
# Import data from seaborn itself
# print(sns.get_dataset_names())

car_crash_df = sns.load_dataset("car_crashes")
# car_crash_df.head()

tips_df = sns.load_dataset("tips")
# tips_df.head()
```

#### Distribution plots

##### Distribution plot

This provides a way to visualize univariate (one variable) distribution

```python
sns.distplot(car_crash_df['not_distracted']) # Plots a histogram and a Kernel Density Estimation (the continuous line on the plot)
```

```python
sns.distplot(car_crash_df["not_distracted"], kde=False) # Plots just the histogram
```

```python
sns.distplot(car_crash_df["not_distracted"], bins=25) # Specifies the number of bins to divide the data into bins to sort of discretize them; ref: https://www.wallstreetmojo.com/data-binning/
```

##### Joint plot

This is used to generate (by default) a scatterplot to compare the distribution of two variables

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crash_df) # Plots a scatterplot with the specified axes values from the specified data source
```

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crash_df, kind="reg") # Plots a scatterplot with the specified axes values from the specified data source and also draws a regression line
```

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crash_df, kind="reg") # Plots a scatterplot with the specified axes values from the specified data source and also draws a regression line for the univariate distributions as well as the bivariate distribution
```

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crash_df, kind="kde") # Plots a KDE contour plot with the specified axes values from the specified data source
```

```python
sns.jointplot(x="speeding", y="alcohol", data=car_crash_df, kind="hex") # Plots a hexagonal plot with the specified axes values from the specified data source
```

##### KDE plot

It is possible to do a Kernel Density Estimation plot directly. Ref: https://towardsdatascience.com/histograms-vs-kdes-explained-ed62e7753f12

```python
sns.kdeplot(car_crash_df["alcohol"])
```

##### Pair plots

These generate scatterplots of every bivariate combination of all of the variables in a dataset

```python
sns.pairplot(car_crash_df)
```

```python
sns.pairplot(tips_df, hue="sex", palette="Blues") # The hue parameter takes in categorical data and assigns different colours to different values of the categorical variable
```

##### Rug plot

A weird way to visualize a univariate distribution. The closer the sticks are, the more dense that particular range of values for the variable

```python
sns.rugplot(tips_df["tip"])
```

#### Styling (1)

Styling plays a key part in how understandable visualizations are

```python
sns.set_style("ticks") # Takes values from white, whitegrid, dark, darkgrid and ticks

plt.figure(figsize=(8,4)) # Sets the size of the following plot

sns.set_context(
    'paper', # Takes values from paper, talk, poster (use as per presentation setting; use talk if to be presented in seminar)
    font_scale=1.2 # Sets font size for labels and axes values
)

sns.jointplot(x="speeding", y="alcohol", data=car_crash_df, kind="reg")

sns.despine(
    left=True, # Gets rid of the spine on the left (and right if right not specified) axis
    bottom=True, # Gets rid of the spine on the bottom (and top if top not specified) axis
)
```

#### Categorical plots

##### Bar plot

These are numerical figures plotted in reference to categorical data

```python
sns.barplot(x="sex", y="total_bill", data=tips_df) # Plots the y-axis values (numerical values) in reference to the x-axis value (categorical values)
```

```python
sns.barplot(x="sex", y="total_bill", data=tips_df, estimator="median") # Plots the y-axis values (numerical values aggregated using the median instead of the mean) in reference to the x-axis value (categorical values); it is also possible to use numpy functions as an estimator
```

##### Count plot

This is similar to a bar plot except that the numerical value is aggregated based on the count function

```python
sns.countplot(x="sex", data=tips_df) # It takes counts of each category of a categorical value
```

##### Box plot

A box plot does a quartile distribution

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=tips_df) # The hue parameter takes the categorical variable that we want to compare across

# More concretely, the following plot shows the !quartile distribution!
# of the !total_bill! both males and females (the values of the
# categorical variable) look after on each !day!. The bar across the
# middle is the median, the whiskers are the min and max values and
# the top and bottom of the boxes are the 25th and 75th percentile
# values. The dots outside are the outliers.
```

##### Violin plot

This is a combination of a box plot and a KDE plot. While a box plot looks at the raw data to generate a quartile-based box, the violin plot uses a KDE to compare between the two categories.

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips_df) # These are uniform KDEs on either side of the split
```

```python
sns.violinplot(x="day", y="total_bill", hue="sex", data=tips_df, split=True) # This does a KDE for each category on either side of the split
```

##### Strip plot

A strip plot is basically a scatter plot that can be separated by values pertaining to each category of a categorical variable

```python
sns.stripplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.stripplot(x="day", y="total_bill", jitter=True, hue="sex", dodge=True, data=tips_df) # jitter sort of separates the points so they do not sit on top of each other, hue separates the point by category of the categorical variable and dodge plots both scatter plots a little separate from each other
```

```python
sns.stripplot(x="day", y="total_bill", jitter=True, hue="sex", data=tips_df) # This is what the scatter plots look like without dodge
```

##### Swarm plot

A swarm plot sort of combines both a violin plot and a strip plot

```python
sns.violinplot(x="day", y="total_bill", data=tips_df, color="white")
sns.swarmplot(x="day", y="total_bill", data=tips_df)
```

```python
sns.violinplot(x="day", y="total_bill", data=tips_df, hue="sex", split=True) # Separating by hue sort of makes the visualization very stupid
sns.swarmplot(x="day", y="total_bill", data=tips_df, hue="sex")
```

#### Matrix Plots

#### Heatmaps

Prints data in a matrix format. It is particularly useful for printing correlations.

```python
plt.figure(figsize=(8,5))
sns.set_context('paper', font_scale=1.5)

sns.heatmap(data=car_crash_df.corr(), annot=True, cmap="Blues") # data has to be in matrix form (.corr() finds correlation between different columns in dataframe, annot=True sets annotation to True and places numbers in the boxes, cmap defines the palette)
```

```python
# It is also possible to create a matrix using the pivot_table function

flights_df = sns.load_dataset("flights")
flights_df = flights_df.pivot_table(index="month", columns="year", values="passengers") # Pivots such that the row numbers are essentially months, the columns are the years and each individual cell would be the number of passengers
sns.heatmap(flights_df, cmap="Blues", linecolor="white", linewidths=1) # linecolor sets the colour of the separating line and linewidths defines the width of the line
```

#### Cluster map

Clusters similar datapoints together across both axes (??)

```python
sns.clustermap(data=flights_df, cmap="Blues")
```

#### Pair grid

A pair grid gives us more fine-grained control of what are present on the diagonals, the upper part and the lower part of the pair plot

```python
# Load Iris dataset
iris_df = sns.load_dataset("iris")
```

```python
iris_pair_grid = sns.PairGrid(data=iris_df, hue="species") # Creates pair grid based on iris dataset
iris_pair_grid.map_diag(plt.hist) # Configures the diagonal elements to be histograms
iris_pair_grid.map_upper(plt.scatter) # Configures the upper part ot be scatterplots
iris_pair_grid.map_lower(sns.kdeplot) # Configures the lower part to be KDE plots
```

#### Regression plots

```python
plt.figure(figsize=(8,6))
sns.set_context("paper", font_scale=1.5)
sns.lmplot(x="total_bill", y="tip", hue="sex", data=tips_df, markers=["^", "+"]) # lmplot plots a linear regression plot; markers defines the style of the points on the plot
```

```python
# It is also possible to do a grid of linear regression plots
sns.lmplot(x="total_bill", y="tip", col="sex", row="time", palette="Set1", data=tips_df)
```
